## Image data augmentation

We increase the training dataset's size by generating modified versions of the existing images. 
By introducing variations to the images, data augmentation helps models generalize more effectively and improve their ability to recognize and interpret new images.

### Import  and initialise ImageDataGenerator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
augmented_image_data = ImageDataGenerator(rotation_range=20,
                                   width_shift_range=0.10, 
                                   height_shift_range=0.10,
                                   shear_range=0.1,
                                   zoom_range=0.1,
                                   horizontal_flip=True,
                                   vertical_flip=True,
                                   fill_mode='nearest',
                                   rescale=1./255
                              )
```

Each set is then augmented and the target classes confirmed:

### Train set augmentation

```python
batch_size = 16 # Set batch size
train_set = augmented_image_data.flow_from_directory(train_path,
                                              target_size=image_shape[:2],
                                              color_mode='rgb',
                                              batch_size=batch_size,
                                              class_mode='binary',
                                              shuffle=True
                                              )

train_set.class_indices
```

### Validation set augmentation

```python
validation_set = ImageDataGenerator(rescale=1./255).flow_from_directory(val_path,
                                                          target_size=image_shape[:2],
                                                          color_mode='rgb',
                                                          batch_size=batch_size,
                                                          class_mode='binary',
                                                          shuffle=False
                                                          )

validation_set.class_indices
```

### Test set augmentation

```python
test_set = ImageDataGenerator(rescale=1./255).flow_from_directory(test_path,
                                                    target_size=image_shape[:2],
                                                    color_mode='rgb',
                                                    batch_size=batch_size,
                                                    class_mode='binary',
                                                    shuffle=False
                                                    )

test_set.class_indices
```

Then each augmented set can be visualised in image plots:

### Augmented train image plot

```python
for _ in range(3):
    img, label = train_set.next()
    print(img.shape)   #  (1,256,256,3)
    plt.imshow(img[0])
    plt.show()
```

### Augmented validation image plot

```python
for _ in range(3):
    img, label = validation_set.next()
    print(img.shape)   #  (1,256,256,3)
    plt.imshow(img[0])
    plt.show()
```

### Augmented test image plot

```python
for _ in range(3):
    img, label = test_set.next()
    print(img.shape)   #  (1,256,256,3)
    plt.imshow(img[0])
    plt.show()
```

# Machine learning (ML) model

---

### Import packages

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Flatten, Dense, Conv2D, MaxPooling2D
```

### Create Basic Model

### Choosing an Activation Function for Binary Classification

* According to [Wikipedia](https://en.wikipedia.org/wiki/Computer_vision#Recognition), "Currently, the best algorithms for such tasks are based on convolutional neural networks.
* **add some more information here**

```python
def create_basic_model(opt, batch_size):
    basic_model = Sequential()

    # Well intially add a single convolution 2D layer with 16 filters, 
    # a kernel size of 3x3, the input size set to our image_shape diemsniosn, 
    # 256x256 and 3 bytes of colour, and the activation as ReLU
    basic_model.add(Conv2D(filters=16,kernel_size=(3,3),input_shape=image_shape,activation='relu'))
    basic_model.add(MaxPooling2D(pool_size=(2,2)))

    # We'll add a second convolution 2D layer with 32 filters
    basic_model.add(Conv2D(32,(3,3),activation='relu'))
    basic_model.add(MaxPooling2D(pool_size=(2,2)))

    # The convolution layer is flattened and fed into a dense, fully-connected layer
    basic_model.add(Flatten())
    basic_model.add(Dense(64,activation='relu'))

    # A single output neuron contiaing a value from 0-1 where 0 = 'healthy' and 1 = 'powdery_mildew' 
    # Sigmoid is chosen as the activation function for Binary Classification
    basic_model.add(Dense(1,activation='sigmoid'))
    
    # Compile the model
    basic_model.compile(loss='binary_crossentropy',
                  optimizer=opt,
                  metrics=['accuracy'])
    
    return basic_model
```

Set optimizer valiable as **adam**

```python
opt = 'adam'
```

```python
create_basic_model(opt, batch_size).summary()
```

### Basic Model Training

We train the model with **Adam** as optimizer for **10 epochs**.  

**Due to frequent IDE (Codeanywhere) disconnects during lengthy sessions, epochs have been limited.  We've attempted to increase the RAM buffer size as described in [this article](https://visualstudiomagazine.com/articles/2022/10/14/binary-classification-using-pytorch-2.aspx#:~:text=The%20two%20most%20common%20are,SGD%20for%20deep%20neural%20networks.), but the problem persists. Ideally we would have trained the model for 20 or more epochs.**

```python
basic_model = create_basic_model(opt, batch_size)
```

```python
basic_model_history = basic_model.fit(
        train_set, # pass in the training set
        steps_per_epoch=len(train_set.classes)//batch_size,
        epochs=10,
        validation_data=validation_set, # pass in the validation set
        verbose=1
        )

# save the model
basic_model.save('outputs/v1/mildew_detector_basic_model.h5')
```

We'll plot the learning accuracies

```python
import matplotlib.pyplot as plt
sns.set_palette("Set2")
sns.set_style("whitegrid")

losses = pd.DataFrame(basic_model.history.history)

losses[['loss','val_loss']].plot(style='.-')
plt.title("Loss")
plt.savefig(f'{file_path}/basic_model_training_losses.png', bbox_inches='tight', dpi=150)
plt.show()

print("\n")
losses[['accuracy','val_accuracy']].plot(style='.-')
plt.title("Accuracy")
plt.savefig(f'{file_path}/basic_model_training_acc.png', bbox_inches='tight', dpi=150)
plt.show()
```

We note that the accuracy and loss figures improve quickly over the first few epochs but show little improvement thereafter.  In light of this we decide to tweak the model's architecture in the following ways: 

* Add a dropout layer to prevent overfitting
* Lengthen epochs to 20 while adding an early stopping callback to stop the training when val_loss doesn't improve for 3 epochs
* Add an additional convolution layer if needed

### Create Model With Tweaked Hyperparameters

```python
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
def create_model(opt, batch_size):
    model = Sequential()
    
    model.add(Conv2D(filters=16,kernel_size=(3,3),input_shape=image_shape,activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Conv2D(32,(3,3),activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    # We'll add a third convolution 2D layer accompanied by maxpooling
    model.add(Conv2D(64,(3,3),activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Flatten())
    model.add(Dense(64,activation='relu'))
    
    # Add dropout to prevent overfitting
    model.add(Dropout(0.25))

    model.add(Dense(1,activation='sigmoid'))
    
    # Compile model
    model.compile(loss='binary_crossentropy',
                  optimizer=opt,
                  metrics=['accuracy'])
    
    return model
```

### Model summary

```python
create_model(opt, batch_size).summary()
```

## Model Training

We'll add an early stopping callback to stop the model when val_loss shows no improvement for 3 epochs

```python
callback = EarlyStopping(monitor='val_loss',patience=3)
```

```python
model = create_model(opt, batch_size)
```

```python
model.fit(
        train_set, # pass in the training set
        steps_per_epoch=len(train_set.classes)//batch_size,
        epochs=20,
        validation_data=validation_set, # pass in the validation set
        callbacks=[callback],
        verbose=1
        )
```

```python
model.save('outputs/v1/mildew_detector_model.h5')
```
